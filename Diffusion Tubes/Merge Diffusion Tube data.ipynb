```python
import pandas as pd
import fuzzymatcher
```

```python
locs = pd.read_csv('Current Tubes.csv')
measurements = pd.read_excel('2017-diffusion-tube-results-summary-reordered_tcm63-392929.xlsx', header=1)
```

```python
measurements['ADDRESS'] = measurements['ADDRESS'].str.upper()
```

```python
locs['Name'] = locs['Name'].str.upper()
```

```python
merged = pd.merge(measurements, locs, how='left', left_on='ADDRESS', right_on='Name')
```

## Fuzzy matching

```python
linktable = fuzzymatcher.link_table(measurements, locs, left_on = "ADDRESS", right_on = "Name", )
```

```python
linktable.to_csv('Link Table.csv')
```

```python
new_linktable = pd.read_csv('Link Table - Manually Edited.csv')
```

```python
with_locs = pd.merge(new_linktable, locs, on='Name')
```

```python
with_mmts = pd.merge(with_locs, measurements, on='ADDRESS')
```

```python
with_mmts.to_csv('DiffusionTubeMeasurements_MergedAndFixed.csv')
```
